```python
import json
import os
import pandas as pd
from utils import get_tables
```

```python
outjson = "./out/json/"
datajson = "./data/ReportJson/"
```

```python
for file in os.listdir(outjson): 
    filebase = file[:-13]
    with open(f"./out/json/{file}", 'r') as f: 
        out = json.load(f)
    with open(f"{datajson}{filebase}.json") as f: 
        docling_json = json.load(f)
    tables = out['tables']
    per_keyword_table_ref = tables["per keyword table refs"]
    for keyword, table_ref in per_keyword_table_ref.items(): 
        with pd.ExcelWriter(f"./out/per_cat_xlsx/{filebase}_{keyword}.xlsx") as writer: 
            for t in table_ref:
                index = int(t.split("/")[-1])
                relevant_table = docling_json['tables'][index]
                df = get_tables.extract_table_to_df(relevant_table) 
                df.to_excel(writer, sheet_name=f"Sheet_table_ref_{index}", index=False)
```
